```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('movie_dataset.csv')
movies
```

```python
movies.info()
```

### we will keep genres, id, keywords, title, overview, cast, crew and  director

```python
movies = movies[['id', 'title', 'genres','keywords', 'overview', 'cast', 'director']]
movies
```

```python
movies.isnull().sum()
```

```python
movies =  movies.fillna('')
movies.isnull().sum()
```

```python
features = ['keywords','cast','genres','overview','director']
def create_tags(df):
    return df['keywords']+" "+df['cast']+" "+df['genres']+" "+df['overview'] +" "+df['director']

movies['tags'] = movies.apply(create_tags, axis=1) 
movies['tags'][0]
```

```python
movies['tags'] = movies['tags'].apply(lambda x:x.lower())
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(stop_words='english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movies_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movies_index]
    movies_list = sorted(list(enumerate(similarity[movies_index])),reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies.iloc[i[0]].title)
    
    
recommend('Avatar')
```

```python
recommend("John Carter")
```
